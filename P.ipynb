```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 假设你的数据存储在一个名为"data.csv"的文件中，特征列名为"feature"，标签列名为"label"
df = pd.read_excel("D:\data\联川\data.xlsx")

# 创建一个新的DataFrame，用于存储每个类别的数据和P值
new_df = pd.DataFrame(columns=["Feature", "Label", "P Value"])

# 遍历每一行数据
for index, row in df.iterrows():
    # 如果该行的标签不是第一或第三类，则添加到新的DataFrame中
    if row["label"] != 0 and row["label"] != 2:
        new_df = new_df.append({"Feature": row["feature"], "Label": row["label"], "P Value": None}, ignore_index=True)
        # 对于每个类别，计算其与其他类别的P值，并将其添加到新的DataFrame中
        for i in [0, 2]:
            if i != row["label"]:
                t_stat, p_val = stats.ttest_ind(df[df["label"] == row["label"]]["feature"], df[df["label"] == i]["feature"])
                new_df = new_df.append({"Feature": row["feature"], "Label": row["label"], "P Value": p_val}, ignore_index=True)
                new_df = new_df.append({"Feature": row["feature"], "Label": i, "P Value": p_val}, ignore_index=True)

# 将特征列转换为分类变量
new_df["Feature"] = new_df["Feature"].astype("category")
new_df["Label"] = new_df["Label"].astype("category")

# 绘制箱线图和P值
plt.figure(figsize=(10, 6))
sns.boxplot(x="Feature", y="Label", data=new_df)
for index, row in new_df.iterrows():
    if not pd.isnull(row["P Value"]):
        plt.text(row["Feature"], row["Label"], f"p={row['P Value']:.4f}", color='red')
plt.show()
```

```python
import numpy as np
from scipy import stats

data = np.random.randn(10,100) #产生10行100列矩阵,数据服从标准正态分布
labels = np.random.randint(0,3,100) #每列随机产生0-2之间的分类标签

pvalues = np.empty((3,3,100)) #保存每一列中不同类之间的P值,三维数组

for i in range(100):
    col_data = data[:,i] #获取第i列数据
    col_labels = labels[i] #获取第i列标签
    
    for c1 in range(3):
        for c2 in range(3):
            if c1 != c2:
                t1 = col_data[col_labels==c1] #类1数据
                t2 = col_data[col_labels==c2] #类2数据
                pvalues[c1,c2,i],_ = stats.ttest_ind(t1,t2) #计算P值

print(pvalues)
```

```python
import pandas as pd
from scipy import stats

# 读取数据
data = pd.read_excel("D:\\data\\联川\\1.xlsx")

# 将数据按照特征进行分组
grouped_data = data.groupby('NUDT4B')

# 对每个组进行t检验，并保存结果到字典中
p_values = {}
for name, group in grouped_data:
    t_statistic, p_value = stats.ttest_ind(group[0], group[1])  # 注意此处应该使用索引0和1来获取每组的两个数值
    p_values[name] = {'t_statistic': t_statistic, 'p_value': p_value}
    # ... 继续添加其他特征的t检验结果

# 输出t检验结果
print(p_values)
```

# 每个特征在不同类两两之间的P值

```python
import numpy as np
import pandas as pd
from scipy import stats

df=pd.read_excel("D:\\data\\联川\\1.xlsx")
data = pd.DataFrame(df.drop(["Symbol","subtype"],axis=1)) 

label = df["subtype"]
labels=np.array(label).astype(np.int32)

p_values = []
for i in range(data.shape[1]):
    t_stat, p_value = stats.ttest_ind(data[labels==1][i],data[labels==2][i])
    p_values.append(p_value)
    
print(p_values)
```
